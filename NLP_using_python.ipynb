```python
import nltk
nltk.download('punkt')
```

```python
for _ in dataset['v2']:
    corpus = str(_)
    break;
print(corpus)
```

```python
from nltk.tokenize import sent_tokenize
sent_tokenize(corpus,language="english")

## treating sentence as a token
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)

## treating words as a token
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
```

```python
word=tokenizer.tokenize(corpus)
word
```

```python
## treebankWordTokenizer doesn't consider fullstop as a word
```

## Proter Stemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for i in word:
    print(i+"---->"+stemming.stem(i))
```

## RegexpStemmer

```python
from nltk.stem import RegexpStemmer
```

```python
regexp=RegexpStemmer("ing$|es$",min=4)
```

```python
regexp.stem('eating')
```

```python
regexp.stem('ingeating')
```

## Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
sbs = SnowballStemmer("english",)
```

```python
for i in word:
    print(i+"----->"+sbs.stem(i))
```

```python
### true difference between snowballstemmer and Regexpstemmer
```

In short, SnowballStemmer is more structured and language-specific, while RegexpStemmer offers flexibility through regular expressions.

```python
## for chatbots this techniches alone is not effective so we move on with lemmatization 
```

```python
## lemmatization ----> More precise, produces valid words.
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
print(lemmatizer.lemmatize("going")) ## pos default is noun
print(lemmatizer.lemmatize("going",pos='v')) ## pos set to verb
```
